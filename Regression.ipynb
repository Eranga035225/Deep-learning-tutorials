```python
import tensorflow as tf
```

```python
# create a model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(4,input_shape=[1]),
    tf.keras.layers.Dense(1)
])
# compile the model
model.compile(loss = tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.Adam(learning_rate=0.01),
              metrics = ['mae'])

# train the model
epoch_number = 10

```
