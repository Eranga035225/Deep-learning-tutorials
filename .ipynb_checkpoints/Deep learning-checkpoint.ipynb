## Sigmoid Function

```python
import numpy as np


def sigmoid(x):
    return 1.0/(1.0 + np.exp(-x))

inputs = [13, 3, -12]

for x in inputs:
    print('{} --> {}'.format(x, sigmoid(x)))

```

## Tanh functions

```python
import numpy as np

def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

inputs = [13, 3, -12]

for x in inputs:
    print('{} --> {}'.format(x, tanh(x)))
```

## ReLU Function


```python
def relu(x):
    return max(0.0,x)

inputs = [13, 3, -12]

for x in inputs:
    print('{} --> {}'.format(x, relu(x)))

```

## Leaky ReLu Function

```python
def leakyrelu(x):
    return max(0.1*x,x)

inputs = [13, 3, -12]

for x in inputs:
    print('{} --> {}'.format(x, leakyrelu(x)))
```

## softmax function 

```python
import numpy as np


def softmax(x):
    return np.exp(x) / np.exp(x).sum()

inputs = [1, 3, -2]

print(softmax(inputs))
```
